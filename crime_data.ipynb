```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('crime_data.csv')
df
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
df_norm = StandardScaler().fit_transform(df)
df_norm
```

## Model Building

```python
db = DBSCAN()
db.fit_predict(df_norm)
```

```python
sns.pairplot(df.iloc[:,0:12])
```

```python
# DBSCAN Clustering

dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(df_norm)
```

```python
# Noisy samples are given the label -1
dbscan.labels_
```

```python
# Adding clusters to dataset
df['cluster']=dbscan.labels_
df
```

```python
df.groupby('cluster').agg(['mean']).reset_index()
```

## KMeans Cluster

```python
km = KMeans()

cluster = km.fit_predict(df_norm)
cluster
```

```python
wcss = []
for i in range(1,10):
    km = KMeans(n_clusters=i)
    y = km.fit_predict(df_norm)
    wcss.append(km.inertia_)
wcss
```

```python
plt.plot(range(1,10),wcss)
plt.xlabel('k-value')
plt.xlabel('k-value')
plt.ylabel('wcss')
```
